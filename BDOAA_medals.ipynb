```python
# Configure font and text settings
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
plt.rcParams.update({
    'font.family': 'serif',
    'font.size': 12
})
```

### Necessary Library

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

## *Additional Bangladesh Flag*

```python
# Define the flag dimensions (keeping 10:6 aspect ratio)
flag_width = 10
flag_height = 6

# Define the red circle's properties
circle_radius = flag_height * 0.3  # 1/5th of the height
circle_x = flag_width * 0.5 - flag_width * 0.05  # Slightly left from center (1/20th of width)
circle_y = flag_height * 0.5  # Vertically centered

# Create figure and axis
fig, ax = plt.subplots(figsize=(flag_width / 2, flag_height / 2))  # Scaled for visibility

# Draw green rectangle (Flag Background)
flag = patches.Rectangle((0, 0), flag_width, flag_height, linewidth=1, edgecolor='black', facecolor='#006A4E')
ax.add_patch(flag)

# Draw red circle (Centered but slightly left)
circle = patches.Circle((circle_x, circle_y), circle_radius, color='red')
ax.add_patch(circle)

# Set plot limits
ax.set_xlim(0, flag_width)
ax.set_ylim(0, flag_height)
ax.set_aspect(1)  # Ensure equal scaling

# Remove axes
ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)

# Show flag
plt.show()
```

# **Medal Data**

```python
# Data
years = [2018, 2019, 2020, 2021, 2022, 2023, 2024]
gold_counts = [0, 0, 0, 0, 0, 0, 0]  # Number of Golds for each year
silver_counts = [0, 0, 2, 0, 1, 2, 1]  # Number of Silvers for each year
bronze_counts = [0, 0, 3, 1, 5, 1, 2]  # Number of Bronzes for each year
hm_counts = [1, 0, 1, 4, 0, 0, 2]  # Number of HMs for each year
```

### Plot without flag

```python
# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the bars
hm_bar = ax.bar(years, hm_counts, label='HM (Honorable Mention)', color='blue')
bronze_bar = ax.bar(years, bronze_counts, label='Bronze', color='#CD7F32', bottom=hm_counts)
silver_bar = ax.bar(years, silver_counts, label='Silver', color='silver',
                    bottom=[h + b for h, b in zip(hm_counts, bronze_counts)])
gold_bar = ax.bar(years, gold_counts, label='Gold', color='gold',
                  bottom=[h + b + s for h, b, s in zip(hm_counts, bronze_counts, silver_counts)])

# Labels and legend
ax.set_xlabel('Year')
ax.set_ylabel('Medal Count')
ax.set_title('Medal Distribution Over Years in IOAA, GeCAA, and OWAO')

# Set legend with correct order
ax.legend(handles=[gold_bar, silver_bar, bronze_bar, hm_bar],
          labels=['Gold', 'Silver', 'Bronze', 'HM (Honorable Mention)'], title='Medal')

ax.set_xticks(years)
ax.set_yticks(range(max(hm_counts) + max(bronze_counts) + max(silver_counts) + max(gold_counts) + 1))

ax.grid(axis='y', linestyle='--', alpha=0.7)

# --- Add Bangladesh Flag in the Upper Left Corner ---
flag_width = 1.5  # Adjust width relative to plot
flag_height = flag_width * (6/10)  # Maintain 10:6 ratio

# Position inside the plot (normalized coordinates)
flag_x = -1  # Adjust position to be inside the axis
flag_y = max(hm_counts) + max(bronze_counts) + max(silver_counts) + max(gold_counts) - 1  # Below the title

# Draw Green Rectangle (Flag Background)
flag = patches.Rectangle((flag_x, flag_y), flag_width, flag_height,
                         linewidth=1, edgecolor='black', facecolor='#006A4E', transform=ax.transData)
ax.add_patch(flag)

# Draw Red Circle (Flag Center)
circle_x = flag_x + flag_width * 0.5  # Center horizontally
circle_y = flag_y + flag_height * 0.5  # Center vertically
circle_radius = flag_width * 0.2  # Adjust radius
flag_circle = patches.Circle((circle_x, circle_y), circle_radius, color='red', transform=ax.transData)
ax.add_patch(flag_circle)

# Show the plot
plt.show()
```

### Plot with Flag

```python
# Create the figure and axes
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the bars
hm_bar = ax.bar(years, hm_counts, label='HM (Honorable Mention)', color='blue')
bronze_bar = ax.bar(years, bronze_counts, label='Bronze', color='#CD7F32', bottom=hm_counts)
silver_bar = ax.bar(years, silver_counts, label='Silver', color='silver',
                    bottom=[h + b for h, b in zip(hm_counts, bronze_counts)])
gold_bar = ax.bar(years, gold_counts, label='Gold', color='gold',
                  bottom=[h + b + s for h, b, s in zip(hm_counts, bronze_counts, silver_counts)])

# Labels and legend
ax.set_xlabel('Year')
ax.set_ylabel('Medal Count')
ax.set_title('Bangladesh Medal Distribution Over Years in \n IOAA, IOAAjr, GeCAA, and OWAO')

# Set legend with correct order
ax.legend(handles=[gold_bar, silver_bar, bronze_bar, hm_bar],
          labels=['Gold', 'Silver', 'Bronze', 'HM (Honorable Mention)'], title='Medal')

ax.set_xticks(years)
ax.set_yticks(range(max(hm_counts) + max(bronze_counts) + max(silver_counts) + max(gold_counts)))
ax.grid(axis='y', linestyle='--', alpha=0.7)

# --- Add Bangladesh Flag ---
flag_height = 2  # Flag should span 2 medal counts
flag_width = flag_height * (7 / 8)  # Maintain the 10:6 ratio

# Position the flag so it starts from 2018 on the x-axis
flag_x = years[0] - 0.1  # Align with 2018
flag_y = 7.5  # Slightly above bars

# Draw Green Rectangle (Flag Background)
flag = patches.Rectangle((flag_x, flag_y), flag_width, flag_height,
                         linewidth=1, edgecolor='black', facecolor='#006A4E', transform=ax.transData, zorder=3)
ax.add_patch(flag)

# Draw Red Circle (Centered but slightly left)
circle_x = flag_x + flag_width * 0.5 - flag_width * 0.05  # Slightly left from center
circle_y = flag_y + flag_height * 0.5  # Centered vertically
circle_width = flag_height * 0.375  # Adjusted width
circle_height = flag_height * 0.65  # Adjusted height (equal for a perfect circle)

# Draw Red Ellipse (to correct aspect ratio)
flag_circle = patches.Ellipse((circle_x, circle_y), circle_width, circle_height, color='red', transform=ax.transData, zorder=4)
ax.add_patch(flag_circle)
ax.add_patch(flag_circle)

# Show the plot
plt.show()
```
